```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
story=[]
list=['/kaggle/input/game-of-thrones-books/004ssb.txt',
      '/kaggle/input/game-of-thrones-books/005ssb.txt',
      '/kaggle/input/game-of-thrones-books/001ssb.txt',
      '/kaggle/input/game-of-thrones-books/002ssb.txt',
      '/kaggle/input/game-of-thrones-books/003ssb.txt']
for item in list:
    f = open(item, encoding='utf-8', errors='ignore')
    corpus=f.read()
    raw_sent=sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

**Creating the Word2Vec model**

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4
)
```

```python
model.build_vocab(story)
```

**Training**

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

## **Dimensionality Reduction using PCA**

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
X=pca.fit_transform(model.wv.get_normed_vectors())
```
